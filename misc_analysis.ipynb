```python
stop_term['Northern Terminus (Stop ID)'] = stop_term['Northern Terminus (Stop ID)'].str.split('|')
stop_term['Southern Terminus (Stop ID)'] = stop_term['Southern Terminus (Stop ID)'].str.split('|')
```

```python
stop_term = stop_term.explode('Northern Terminus (Stop ID)')
stop_term = stop_term.explode('Southern Terminus (Stop ID)')
```

```python
final = pd.merge(stop_term, stops[['stop_id', 'stop_lon', 'stop_lat']], left_on='Northern Terminus (Stop ID)', right_on='stop_id')
final = pd.merge(final, stops[['stop_id', 'stop_lon', 'stop_lat']], left_on='Southern Terminus (Stop ID)', right_on='stop_id', suffixes=('_north', '_south')) 
final.drop(['stop_id_north', 'stop_id_south'], inplace=True, axis=1)

```

```python
final = pl.from_pandas(final)
```

```python
shapes['Line'] = shapes['shape_id'].apply(lambda x: re.match(r"^(\w+)\.", x).groups(1)[0])
```

```python
import polars as pl


# Filter to get the beginning coordinates (point_num 0)
begin_df = shapes_pl.filter(pl.col("shape_pt_sequence") == 0).rename({
    "shape_pt_lat": "begin_lat",
    "shape_pt_lon": "begin_lon"
})

# Filter to get the ending coordinates (max point_num for each key)
max_point_num_df = shapes_pl.group_by("shape_id").agg(pl.col("shape_pt_sequence").max().alias("max_point_num"))
end_df = shapes_pl.join(max_point_num_df, left_on=["shape_id", "shape_pt_sequence"], right_on=['shape_id', 'max_point_num']).rename({
    "shape_pt_lat": "end_lat",
    "shape_pt_lon": "end_lon",
    "shape_pt_sequence": "max_point"
})

# Join the beginning and ending coordinates on the key
result_df = begin_df.join(end_df, on="shape_id").select([
    "shape_id",
    "begin_lat",
    "begin_lon",
    "max_point",
    "end_lat",
    "end_lon"
])

result_df.head()
```

```python
result_df = result_df.with_columns(
    pl.concat_str(['begin_lat', 'begin_lon', 'end_lat', 'end_lon'], separator=',').alias('concatenation')
)
```
